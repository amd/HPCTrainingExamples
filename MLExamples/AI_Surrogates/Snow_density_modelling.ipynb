```python
import numpy as np
import matplotlib.pyplot as plt

# Constants
rho_min = 100  # Minimum snow density (kg/m)
T_ref = 0  # Reference temperature for density adjustment (C)

# Scaling factors for temperature, wind speed, and humidity
alpha_T = 1.5   # Density increase per degree C below reference temperature
alpha_u = 10    # Density increase per m/s wind speed
alpha_h = 0.5   # Density increase per % humidity

# Function to calculate snow density
def calculate_snow_density(T, u, h):
    """
    Calculate snow density based on temperature, wind speed, and humidity.
    
    Parameters:
    T: Temperature (C)
    u: Wind speed (m/s)
    h: Humidity (%)
    
    Returns:
    Snow density (kg/m)
    """
    # Snow density formula
    density = rho_min + alpha_T * (T_ref - T) + alpha_u * u + alpha_h * h
    # Ensure density stays within a reasonable range (e.g., 100 to 600 kg/m)
    return np.clip(density, 100, 600)

# Example parameters
temperatures = np.linspace(-20, 0, 100)  # Range of temperatures (-20C to 0C)
wind_speed = 5  # Constant wind speed (m/s)
humidity = 80   # Constant relative humidity (%)

# Calculate snow density for varying temperatures
snow_densities = calculate_snow_density(temperatures, wind_speed, humidity)

# Plot the result
plt.figure(figsize=(8, 6))
plt.plot(temperatures, snow_densities, label=f'Wind Speed = {wind_speed} m/s, Humidity = {humidity}%')
plt.title('Effect of Temperature on Snow Density')
plt.xlabel('Temperature (C)')
plt.ylabel('Snow Density (kg/m)')
plt.grid(True)
plt.legend()
plt.show()

# Explore impact of wind speed and humidity variation
wind_speeds = [0, 5, 10]  # m/s
humidity_levels = [50, 80, 100]  # %

# Plot effect of wind speed
plt.figure(figsize=(8, 6))
for u in wind_speeds:
    densities = calculate_snow_density(temperatures, u, humidity)
    plt.plot(temperatures, densities, label=f'Wind Speed = {u} m/s')
plt.title('Effect of Wind Speed on Snow Density')
plt.xlabel('Temperature (C)')
plt.ylabel('Snow Density (kg/m)')
plt.grid(True)
plt.legend()
plt.show()

# Plot effect of humidity
plt.figure(figsize=(8, 6))
for h in humidity_levels:
    densities = calculate_snow_density(temperatures, wind_speed, h)
    plt.plot(temperatures, densities, label=f'Humidity = {h}%')
plt.title('Effect of Humidity on Snow Density')
plt.xlabel('Temperature (C)')
plt.ylabel('Snow Density (kg/m)')
plt.grid(True)
plt.legend()
plt.show()
```
